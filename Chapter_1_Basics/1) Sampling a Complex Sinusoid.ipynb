# Sampling a Complex Sinusoid

---

This notebook demonstrates how to generate and visualize a **complex sinusoidal signal**,  
the basic building block of radar signal processing.

- **Objective:** Show how a continuous wave can be represented digitally using complex samples (I/Q).  
- **Concepts:** Complex exponential, in-phase & quadrature components, sampling, Nyquist frequency.  
- **Equation:**

$$
x(t) = e^{j(2\pi f_0 t + \phi)}
$$

- **Connection to radar:** This represents the baseband form of a radar tone;  
  phase variations later relate to Doppler and range.

---

```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters
fs = 10000.0   # sampling rate (Hz)
T = 0.01        # duration (s)
f0 = 1234.0     # The signals frequency (tone)
phi = np.deg2rad(20)  # initial phase (rad)

t = np.arange(0, T, 1/fs) 
x = np.exp(1j*(2*np.pi*f0*t + phi))

# Plot real/imag parts
plt.figure()
plt.plot(t, x.real, label='Re{x}')
plt.plot(t, x.imag, label='Im{x}')
plt.xlabel('Time (s)'); plt.ylabel('Amplitude'); plt.title('Complex Tone (time)')
plt.legend()
plt.show()
print(f"Num samples = {len(t)}, Nyquist = {fs/2} Hz")
```
