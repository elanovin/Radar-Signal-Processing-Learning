# FFT of a Stationary Target

---

**Aim:**  
To show the FFT of a stationary radar target represented by a **single complex sinusoid** with no Doppler shift.

- **Objective:**  
  Show that a stationary target produces a single sharp spectral peak at its tone frequency.
    
**Key Equations:**

The baseband signal for a stationary target is given by  
$$
x(t) = e^{j(2\pi f_0 t + \phi)}
$$

After sampling with sampling frequency \( f_s \):  
$$
x[n] = e^{j(2\pi f_0 n / f_s + \phi)}, \quad n = 0, 1, 2, ..., N-1
$$

The Discrete Fourier Transform (DFT) converts the signal from the time domain into the frequency domain:  
$$
X[k] = \sum_{n=0}^{N-1} x[n] \, e^{-j(2\pi kn / N)}
$$

which shows a single spectral component (spike) at the normalized frequency  
$$
f = \frac{k f_s}{N} = f_0
$$

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 10000.0   # sampling rate (Hz)
T = 0.01       # duration (s)
f0 = 1500.0    # tone frequency (Hz)
phi = np.deg2rad(20)  # initial phase (rad)

t = np.arange(0, T, 1/fs)
x = np.exp(1j*(2*np.pi*f0*t + phi))
```

```python
# Step 2: FFT and frequency resolution
N = len(x)
X = np.fft.fft(x, n=N)
freq = np.fft.fftfreq(N, d=1/fs)

plt.figure()
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(20*np.log10(np.abs(X))))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude (dB)')
plt.title('FFT Magnitude (no window)')
plt.show()

df = fs/N
print(f"Frequency resolution Δf = {df:.2f} Hz")
```
